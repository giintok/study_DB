**웹 스크래핑**

스팀 게임 추천을 하면서 만들었던 app_id를 이용해서 스팀 긍정 리뷰수, 부정 리뷰수 스크래핑 코드

````python
from bs4 import BeautifulSoup
import requests
import re
import pandas as pd
import time
````

````python
# app_ids.csv 파일에서 app_id 불러오기
appids = pd.read_csv('app_ids.csv')['AppID']

# 결과를 저장할 리스트 생성
results = []

# 각 app_id에 대해 반복
for appid in appids[:1000]:
    # 게임 페이지 URL 생성
    url = f'https://store.steampowered.com/app/{appid}/'

    # GET 요청을 보내고 응답 받기
    response = requests.get(url)

    # 딜레이 추가
    time.sleep(3)

    # BeautifulSoup을 사용하여 HTML 파싱 / 응답받은것중 text(HTML, JSON)만 추출하기 위해서 .text / 바이너리 데이터(이미지, pdf) 는 .content
    soup = BeautifulSoup(response.text, 'html.parser')

    # 리뷰 타입이랑 수를 추출 rows = soup.select('div > ul > li')를 사용하지 않고 element로 직접 바로 들어감
    #<input type="radio" name="review_type" value="positive" id="review_type_positive" onchange="ShowFilteredReviews()">
    #<span class="user_reviews_count">(260,049)</span>
    positive_reviews = soup.find('input', {'id': 'review_type_positive'}).find_next('span', {'class': 'user_reviews_count'}).text
    negative_reviews = soup.find('input', {'id': 'review_type_negative'}).find_next('span', {'class': 'user_reviews_count'}).text

    # 숫자만 추출하여 정수로 변환합니다.
    positive_count = int(''.join(filter(str.isdigit, positive_reviews)))
    negative_count = int(''.join(filter(str.isdigit, negative_reviews)))

    # 결과를 딕셔너리로 저장 및 터미널에서 작동 확인
    result = {'app_id': appid, 'positive_reviews': positive_count, 'negative_reviews': negative_count}
    results.append(result)

    print(result)

df = pd.DataFrame(results)
df.to_csv('review_counts_1000.csv', index=False)
````

**정규화**

1. Min-Max 스케일링 : 주로 각 특성(차원)의 값의 데이터의 범위를 0에서 1 또는 원하는 범위로 변환
   
$$
    정규화된 값 = \frac{{\text{{원래 값}} - \text{{최소값}}}}{{\text{{최대값}} - \text{{최소값}}}}
$$
2. 표준화 : from sklearn.preprocessing import StandardScaler,
    데이터를 평균이 0이고 표준 편차가 1인 분포

$$
    표준화된 값 = \frac{{\text{{원래 값}} - \text{{평균}}}}{{\text{{표준 편차(std)}}}}
$$
3. 소프트맥스 정규화: 다중 클래스 분류(다차원) 데이터의 각 요소를 [0, 1] 범위로 변환


    def softmax(x):

        return exp_x / np.sum(exp_x)

$$
    소프트맥스 정규화된 값 = \frac{{e^{x_i}}}{{\sum_{j=1}^{n} e^{x_j}}}
